## Data Preprocessing

#### Column Transformer

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('pow_transformer', PowerTransformer(), ['Hours_Studied']),
                                     ('scaler', StandardScaler(), ['Attendance', 'Sleep_Hours', 'Previous_Scores']),
                                     ('encoder', OrdinalEncoder(categories=[['Low', 'Medium', 'High'], ['Low', 'Medium', 'High'], ['Yes','No'],
                                                                            ['Low', 'Medium', 'High'], ['No','Yes'], ['Low', 'Medium', 'High'],
                                                                            ['No', 'Yes'],['High School','College','Postgraduate'],
                                                                            ['10','9','8','7','6','5','4'], ['6', '5', '4', '3', '2', '1', '0']]), 
                                      ['Parental_Involvement','Access_to_Resources','Extracurricular_Activities','Motivation_Level','Internet_Access',
                                       'Teacher_Quality','Learning_Disabilities','Parental_Education_Level','Sleep_Hours','Physical_Activity']),
                                     ('onehot', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'), 
                                      ['Family_Income', 'School_Type', 'Peer_Influence', 'Distance_from_Home', 'Gender'])],
                       remainder='passthrough', sparse_threshold=0)
ct
```

## Data Splitting

```python
X = students_df.drop(columns='Exam_Score')
Y = students_df['Exam_Score']
X.shape, Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
X_train.shape, y_train.shape
```

## Model Selection

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor, GradientBoostingRegressor, HistGradientBoostingRegressor

models = {
    'lin_reg': LinearRegression(),
    'svm': SVR(),
    'decision_tree': DecisionTreeRegressor(),
    'knn': KNeighborsRegressor(),
    'random': RandomForestRegressor(),
    'extra_trees': ExtraTreesRegressor(),
    'gbm': GradientBoostingRegressor(),
    'hgbm': HistGradientBoostingRegressor()
}
```

#### Cross Validations

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, mean_squared_error

scores_df = pd.DataFrame(index=models.keys(), columns=['MSE', 'MSE_mean'])
X_pr = ct.fit_transform(X)

for i in models.keys():
    scores = cross_val_score(models[i], X_pr, Y, scoring = make_scorer(mean_squared_error), cv = 5)
    scores_df.loc[i, 'MSE'] = np.round(scores, 2)
    scores_df.loc[i, 'MSE_mean'] = np.round(scores.mean(), 2)
    print(f'{i} --> cross validated')
```

```python
scores_df.sort_values(by='MSE_mean')
```

## Model Building

```python
from sklearn.pipeline import Pipeline
pipe = Pipeline(steps=[('transformer', ct), ('lin_reg', LinearRegression())])
pipe.fit(X_train, y_train)
```

## Model Evaluation

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_log_error, mean_absolute_percentage_error, median_absolute_error, max_error
from sklearn.metrics import explained_variance_score, mean_tweedie_deviance, mean_pinball_loss

metrics = {
    'mean_squared_error': mean_squared_error,
    'r2_score': r2_score,
    'mean_absolute_error': mean_absolute_error,
    'mean_squared_log_error': mean_squared_log_error,
    'mean_absolute_percentage_error': mean_absolute_percentage_error,
    'median_absolute_error': median_absolute_error,
    'max_error': max_error,
    'explained_variance_score': explained_variance_score,
    'mean_tweedie_deviance': mean_tweedie_deviance,
    'mean_pinball_loss': mean_pinball_loss
}

evaluation_df = pd.DataFrame(index=metrics.keys(),columns=['Score_or_Loss'])
y_pred = pipe.predict(X_test)

for i in metrics.keys():
    evaluation_df.loc[i, 'Score_or_Loss'] = metrics[i](y_test, y_pred).round(2)

evaluation_df
```
